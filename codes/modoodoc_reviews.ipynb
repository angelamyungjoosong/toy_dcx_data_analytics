```python
from selenium import webdriver
browser = webdriver.Chrome(executable_path='../../../chromedriver.exe')
browser.get('https://www.modoodoc.com/')
```

### 로그인 

```python
# 로그인 버튼 클릭 
# tag: document > ul > li:nth-child(8) > a
browser.find_element_by_css_selector('#__next > div > div > header > nav > div.styles_rightSideBox__dI2tv > button').click()
browser.implicitly_wait(10)
```

```python
browser.find_element_by_css_selector('div.styles_modalContent__RaR_5 > div > button > span').click()
```

```python
# 로그인 
# username tag: #loginId--1
# password tag: #password--2
# submit button tag: #mainContent > div > div > form > div.confirm_btn > button.btn_g.highlight.submit
browser.find_element_by_css_selector('#loginId--1').send_keys('**')
browser.find_element_by_css_selector('#password--2').send_keys('**')
browser.find_element_by_css_selector('#mainContent > div > div > form > div.confirm_btn > button.btn_g.highlight.submit').click()
```
